```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit

housing_data_frame = pd.read_csv("C:\\Users\\sumit\\Downloads\\housing\\housing\\housing.csv")
```

```python
housing_data_frame
```

```python
housing_data_frame["income_cat"] = pd.cut(housing_data_frame["median_income"],
                                        bins=[0.0, 1.5, 3.0, 4.5, 6.0, np.inf],
                                        labels=[1,2,3,4,5])
```

```python
housing_data_frame
```

```python
strat_train_set = None
strat_test_set = None
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
file_name="housing.csv"
```

```python
for train_index,test_index in split.split(housing_data_frame, housing_data_frame["income_cat"]):
    strat_train_set = housing_data_frame.loc[train_index].drop(["income_cat"],axis=1)
    strat_test_set = housing_data_frame.loc[test_index].drop(["income_cat"],axis=1)

    train_file_path = "D:\\TestFolder\\TestprojectMachineLearning\\Train\\train.csv"

    test_file_path = "D:\\TestFolder\\TestprojectMachineLearning\\Test\\test.csv"
            
    if strat_train_set is not None:
        strat_train_set.to_csv(train_file_path,index=False)

    if strat_test_set is not None:
        strat_test_set.to_csv(test_file_path,index=False)        
```
